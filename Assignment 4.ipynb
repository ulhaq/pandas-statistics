# Task 4

This will show two maps for individuals using the Internet in some European countries

This comparison can be helpful to determine countries which are in the first row with regards to adopting IT, and how the countries have changed their usage of Internet over time.

```python
download_extract('http://api.worldbank.org/v2/en/indicator/IT.NET.USER.ZS?downloadformat=csv')

fig, ax = plt.subplots(ncols=2)
fig.suptitle("Individuals using the Internet in some European countries")

year1 = '2016'
year2 = '2000'
country_codes = ['AUT', 'BEL', 'DEU', 'DNK', 'ESP', 'FIN', 'GBR', 'GRC', 'HUN', 'ITA', 'LUX', 'NLD', 'POL', 'PRT', 'SWE']

ax[0].set_title(f"in {year1}")
ax[1].set_title(f"in {year2}")

it_2016 = get_internet_for_country(year1, country_codes)
it_2005 = get_internet_for_country(year2, country_codes)

it_2016.plot(ax=ax[0], column=year1, legend=True, cmap='OrRd')
it_2005.plot(ax=ax[1], column=year2, legend=True, cmap='OrRd')
```
